```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('fifa.csv',index_col='Date',parse_dates=True)
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(16,5))
ax = sns.lineplot(data=df)
ax.set_title("Plot fifa")
```

```python
df3 = pd.read_csv('insurance.csv')
df3.head()
```

```python
sns.scatterplot(x=df3['bmi'] , y=df3['charges'],palette='mako')
```

```python
sns.regplot(x=df3['bmi'] , y=df3['charges'])
```

```python
sns.scatterplot(x=df3['bmi'] , y=df3['charges'] , hue=df3['smoker'])
```

```python
sns.lmplot(data=df3 , x='bmi',y='charges',hue='smoker')
```

```python
sns.swarmplot(data=df3 , x='region',y='charges',hue='smoker')
```

```python
df4 = pd.read_csv('iris.csv', index_col='Id')
df4.head()
```

```python
sns.histplot(df4['Sepal Length (cm)'])
```

```python
sns.histplot(data=df4 ,x='Sepal Length (cm)', hue='Species' , kde=True)
```

```python
sns.histplot(data=df4 ,x='Sepal Length (cm)', hue='Species' ,palette='mako')
```

```python
sns.histplot(df4['Sepal Length (cm)'] , kde=True)
```

```python
sns.kdeplot(df4['Sepal Width (cm)'] , shade=True)
```

```python
sns.jointplot(data=df4 , x='Sepal Width (cm)',y='Sepal Length (cm)',kind="reg")
```

```python
sns.jointplot(data=df4 , x='Sepal Width (cm)',y='Sepal Length (cm)',kind="kde")
```
