## Scaling and Normalization

```python
from mlxtend.preprocessing import minmax_scaling
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

original_data = np.random.exponential(size=1000)
scaled_data = minmax_scaling(original_data, columns=[0]) 

fig , axis = plt.subplots(1,2,figsize=(15,6))
sns.histplot(original_data , ax=axis[0] , kde=True)
axis[0].set_title('original data')
sns.histplot(scaled_data , ax=axis[1] , kde=True)
axis[1].set_title("scaled data")
plt.show()
```

```python
from scipy import stats
```

```python
normalized_data = stats.boxcox(original_data)

fig , ax = plt.subplots(1,2,figsize=(15,6))
sns.histplot(original_data , kde=True,ax=ax[0])
ax[0].set_title("original data")
sns.histplot(normalized_data , kde=True , ax=ax[1])
ax[1].set_title("normalized data")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
norm_data = scaler.fit_transform(original_data.reshape(-1, 1))
norm_data
```

```python
fig , ax = plt.subplots(1,2,figsize=(15,6))
sns.histplot(norm_data , kde=True,ax=ax[0])
ax[0].set_title("StanderScaler data")
sns.histplot(normalized_data , kde=True , ax=ax[1])
ax[1].set_title("BoxCox data")
```
